```python
import numpy as np
import matplotlib.pyplot as plt

# For point (x,y) and image size h=Image_Hieght-1 and w=Image_Width-1, the pivot is changed to be at the center (x,y) -> (x-w/2,y-h/2).
def moveBy90(height, width, i, j): # When this function executes twice image will be rotated by 180 degrees
    iNew = j
    jNew = -i + height - 1
    return int(iNew), int(jNew)

image1 = plt.imread('rotate.png')
print(image1.shape)
image2 = np.zeros([image1.shape[1], image1.shape[0], image1.shape[2]], dtype=image1.dtype)
for c in range(4):
    for i in range(image1.shape[0]-1):
        for j in range(image1.shape[1]-1):
            newPoint = moveBy90(image1.shape[0], image1.shape[1], i, j)
            newPoint = moveBy90(image1.shape[0], image1.shape[1], newPoint[0], newPoint[1])
            newPoint = moveBy90(image1.shape[0], image1.shape[1], newPoint[0], newPoint[1])
            try:
                image2[newPoint[0], newPoint[1], c]=image1[i,j,c]
            except IndexError as error:
                print(error)

plt.subplot(1, 2, 1)
plt.imshow(image1)

plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.show()
```
